```python
#Se valida si existen registros nulos en el dataframe
for columna in df.columns:
    num_nulos = df[columna].isnull().sum()
    if num_nulos > 0:
        print(f"La columna '{columna}' tiene {num_nulos} valores nulos.")
```

```python
#Se verifican y cambian los valores nulos encontrados en la columna 'RAZ√ìN SOCIAL' por el valor 'No registra'

#Lista de indices que se van a verificar
indices = [3529, 13550]
nuevo_valor = 'No registra'

#Ciclo para verificar y reemplazar los valores nulos de la columna
for idx in indices:
    if pd.isnull(df.loc[idx, 'RAZ√ìN SOCIAL']):
        df.loc[idx, 'RAZ√ìN SOCIAL'] = nuevo_valor

#Ciclo para verificar si los cambios se realizaron
for columna in df.columns:
    num_nulos = df[columna].isnull().sum()
    if num_nulos > 0:
        print(f"La columna '{columna}' tiene {num_nulos} valores nulos.")
    else:
        cambio_realizado = True

#Ciclo para mostrar si el cambio se realiz√≥ o no
if cambio_realizado:
       print("El cambio se realiz√≥ con √©xito‚úÖ.")
else:
    print("No se realizaron cambios‚ùå.")
    
```

```python
#Se visualizan los cambios de valores nulos
print(df.loc[3529])
print("-" * 50)
print(df.loc[13550])
```

```python
#El siguiente paso es eliminar de las observaciones de las columnas los espacios en blanco al inicio y al final
#Adem√°s se cuentan cuantos espacios en blanco se eliminan

def contar_espacios_eliminados(df):
    eliminaciones_totales = 0
    for columna in df.select_dtypes(include=['object']).columns:
        # Se cuentan los espacios antes de la limpieza
        longitud_original = df[columna].str.len()
        # Se aplica la limpieza
        df[columna] = df[columna].str.strip()
        # Se cuentan los espacios despu√©s de la limpieza
        longitud_limpia = df[columna].str.len()
        # Se calcula el n√∫mero de eliminaciones
        eliminaciones = (longitud_original - longitud_limpia).sum()
        eliminaciones_totales += eliminaciones
    return eliminaciones_totales

# Se cuentan los espacios eliminados
espacios_eliminados = contar_espacios_eliminados(df)
print("Espacios eliminados:", espacios_eliminados)
```

```python
#Se observan cuantas categor√≠as tiene cada variable de tipo 'object'

for columna in df.select_dtypes(include=['object']).columns:
    # Se convierten los valores a cadena de texto
    df[columna] = df[columna].astype(str).str.strip()
    
    # Se obtienen las categor√≠as √∫nicas
    categorias = df[columna].unique()
    
    # Se cuentan el n√∫mero de categor√≠as
    num_categorias = len(categorias)
    
    # Se visualiza la informaci√≥n de las categor√≠as por columna
    print(f"Columna: {columna}")
    print(f"N√∫mero de Categor√≠as: {num_categorias}")
    print("-" * 40) 
```

```python
#Se visualizan las categor√≠as de las columnas si tienen 7 o menos

for columna in df.select_dtypes(include=['object']).columns:
    df[columna] = df[columna].astype(str).str.strip()
    categorias = df[columna].unique()
    num_categorias = len(categorias)
    
    #Verificar columnas con 7 categor√≠as o menos para visualizarlas
    if num_categorias <= 7:
        print(f"Columna: {columna}")
        print(f"N√∫mero de Categor√≠as: {num_categorias}")
        print(f"Categor√≠as: {', '.join(categorias)}")
        print("-" * 40)
```

Se suprimen las t√≠ldes en las observaciones para evitar discrepancias en el an√°lisis, comparaciones y agrupamientos, lo que facilita un procesamiento m√°s uniforme y preciso en el tratamiento y modelado de datos. Para esta labor se requiere la biblioteca 'unidecode'

```python
from unidecode import unidecode

#Funci√≥n para eliminar tildes
def eliminar_tildes(texto):
    return unidecode(texto)

# Aplicaci√≥n de la funci√≥n para eliminar t√≠ldes sobre las columnas tipo 'object'
for columna in df.select_dtypes(include=['object']).columns:
    df[columna] = df[columna].apply(eliminar_tildes)

# Funci√≥n para validar t√≠ldes en el dataframe
def tiene_tildes(texto):
    return any(c in '√°√©√≠√≥√∫√º' for c in texto)

# Se comprueba si alguna observaci√≥n tiene t√≠lde
tildes_presentes = False
for columna in df.select_dtypes(include=['object']).columns:
    if df[columna].apply(tiene_tildes).any():
        tildes_presentes = True
        break

if tildes_presentes:
    print("A√∫n hay tildes presentes en el DataFrame‚ùå.")
else:
    print("Se han eliminado todas las tildes del DataFrame‚úÖ.")

```

Adem√°s, se convierten todas las observaciones de texto a min√∫sculas con el mismo fin por el que se eliminaron las tildes.

```python
# Funci√≥n para convertir texto a min√∫sculas
def convertir_minusculas(texto):
    return texto.lower()

# Se aplica la funci√≥n de conversi√≥n a min√∫scula en las columnas tipo 'object'
for columna in df.select_dtypes(include=['object']).columns:
    df[columna] = df[columna].apply(convertir_minusculas)

# Funci√≥n para validar si hay may√∫sculas en el Dataframe
def tiene_mayusculas(texto):
    return any(c.isupper() for c in texto)

# Se comprueba si a√∫n existen may√∫sculas
mayusculas_presentes = False
for columna in df.select_dtypes(include=['object']).columns:
    if df[columna].apply(tiene_mayusculas).any():
        mayusculas_presentes = True
        break

if mayusculas_presentes:
    print("A√∫n hay may√∫sculas presentes en el DataFrame‚ùå.")
else:
    print("Se han convertido todas las observaciones a min√∫sculas en el DataFrame‚úÖ.")

```

Se verifica si alguna de las observaciones de la columna 'NIT' contiene la longitud correcta (9 d√≠gitos)

```python
# Funci√≥n para ajustar la longitud del Nit
def ajustar_nit(nit):
    if pd.isnull(nit):
        return nit  # Si el valor es nulo, no lo modifica
    nit_str = str(nit)  # Se convierte a texto para contar f√°cilmente el n√∫mero de caracteres.
    if len(nit_str) == 9:
        return nit_str  # Si el valor es correcto, se mantiene
    elif len(nit_str) > 9:
        return nit_str[:9]  # Quitar los d√≠gitos existentes si tiene m√°s de 9 d√≠gitos
    else:
        return 'Incorrecto'  # Indicar como incorrecto si tiene menos de 9 d√≠gitos

# Se aplica la funci√≥n a la columna 'NIT'
df['NIT'] = df['NIT'].apply(ajustar_nit)

# Se verifican los cambios realizados
num_incorrectos = df[df['NIT'] == 'Incorrecto'].shape[0]
if num_incorrectos > 0:
    print(f"Se encontraron {num_incorrectos} NITs incorrectos.")
else:
    print("Todos los NITs est√°n correctos o ajustados‚úÖ.")
```

# Actividad 2. Transformaci√≥n de datos üîÅ

```python
#Se importa la libreria 'tabulate' para su uso en el desarrollo del informe
from tabulate import tabulate
```

Para iniciar se visualizan las variables con su tipo de datos:

```python
# C√≥digos de escape ANSI para negrita
NEGRITA = '\033[1m'
RESET = '\033[0m'

# Se obtienen datos para construir la tabla
datos = [(columna, tipo) for columna, tipo in df.dtypes.items()]

# Se imprime el t√≠tulo en negrita
print(f"{NEGRITA}Variables y sus tipos de datos:{RESET}")

# Se imprime la tabla
print(tabulate(datos, headers=['Columna', 'Tipo de Dato'], tablefmt='grid'))
```

Para facilitar un an√°lisis posterior de los datos de las columnas *'INGRESOS OPERACIONALES'*, *'GANANCIA (P√âRDIDA)'*,*'TOTAL ACTIVOS'*, *'TOTAL PASIVOS'* y *'TOTAL PATRIMONIO'* lo ideal es convertirlas en n√∫mericas. Por lo anterior y teniendo en cuenta que estos datos est√°n en billones de pesos, a estos registros se les debe eliminar el signo $ y hacer la respectiva conversi√≥n para cambiar el tipo de datos de las columnas mencionadas.

```python
#Se realiza una lista de las columnas a las que se les quitar√° el signo $
columnas_con_signo = ['INGRESOS OPERACIONALES', 'GANANCIA (P√âRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']

#Ciclo para quitar el signo $
for columna in columnas_con_signo:
    if df[columna].dtype == 'object': 
        df[columna] = df[columna].str.replace('$', '', regex=False)

#Se visualizan las primeras filas del Dataframe para verificar el cambio
df.head()
```

```python
# Se enlistan las columnas que se convertiran a numericas
columnas_a_convertir = ['INGRESOS OPERACIONALES', 'GANANCIA (P√âRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']

# Ciclo para convertir el valor a billones y a variable num√©rica
for columna in columnas_a_convertir:
    if df[columna].dtype == 'object':  
        # Conversi√≥n a variable num√©rica
        df[columna] = pd.to_numeric(df[columna], errors='coerce')
    # Conversi√≥n a billones
    df[columna] = df[columna] * 1e9

# Se visualiza las primeras filas del Dataframe para verificar el cambio
df.head()
```

Se convierte la variable 'NIT' a tipo *object* porque esta se utiliza principalmente para identificaci√≥n y no para c√°lculos num√©ricos. Esto evita errores de interpretaci√≥n, preserva el formato y asegura que todas las operaciones sobre esta variable se realicen correctamente.
Por la misma raz√≥n se cambia el tipo de la variable CIIU, ya que este es un indentificador de categor√≠as industriales.

```python
# Se convierten las variables 'NIT' y 'CIIU' a tipo 'object'
df['NIT'] = df['NIT'].astype('object')
df['CIIU'] = df['CIIU'].astype('object')

#Se verifica el tipo de dato de las variables 'NIT' y 'CIIU'
df[['NIT', 'CIIU']].dtypes
```
